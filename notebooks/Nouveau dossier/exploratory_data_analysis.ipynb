```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("Set1")
```

**Loading Dataset**

```python
train_identity = pd.read_csv('../data/raw/train_identity.csv')
train_transaction = pd.read_csv('../data/raw/train_transaction.csv')
test_identity = pd.read_csv("../data/raw/test_identity.csv")
test_transaction = pd.read_csv("../data/raw/test_transaction.csv")
data1 = pd.read_csv("../data/raw/data1.csv")
dataset2 = pd.read_csv("../data/raw/creditcard.csv")
# combining the data
train = pd.merge(train_transaction, train_identity, on='TransactionID', how='left')
test = pd.merge(test_transaction, test_identity, on='TransactionID', how='left')
```

```python
print(f'Train dataset has {train.shape[0]} rows and {train.shape[1]} columns.')
print(f'Test dataset has {test.shape[0]} rows and {test.shape[1]} columns.')
```

```python
# Visualizing missing values
plt.figure(figsize=(10, 6))
sns.heatmap(train.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Target variable distribution
plt.figure(figsize=(6, 4), palette="pastel")
sns.countplot(x="isFraud", data=train)
plt.title("Distribution of Target Variable (isFraud)")
plt.xlabel("isFraud")
plt.ylabel("Count")
plt.show()

# Percentage of fraudulent vs non-fraudulent transactions
fraud_percentage = train["isFraud"].value_counts(normalize=True) * 100
print("Fraud Percentage:\n", fraud_percentage)
```
